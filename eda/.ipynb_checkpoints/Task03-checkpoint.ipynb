### 5.1数据合并 ###

```python
import numpy as np
import pandas as pd

#5.1.1载入数据，观察数据表之间的关系
text_left_up = pd.read_csv('data/train-left-up.csv')
text_left_down =  pd.read_csv('data/train-left-down.csv')
text_right_up = pd.read_csv('data/train-right-up.csv')
text_right_down = pd.read_csv('data/train-right-down.csv')

print(len(text_left_up))
print(len(text_left_down))
print(len(text_right_up))
print(len(text_right_down))
```

```python
text_left_up.head()
```

```python
text_left_down.head()
```

```python
text_right_up.head()
```

```python
text_right_down.head()
```

#### pd.concat 默认的结合方式为outer ####

```python
#5.1.2 使用concat方法，将数据train-left-up.csv和train-right-up.csv横向合并为一张表，并保存为result-up
result_up = pd.concat([text_left_up, text_right_up], axis=1,join='inner')
result_up.tail()
```

```python
#5.1.3 使用concat方法，将数据train-left-down.csv和train-right-down.csv横向合并为一张表，并保存为result-down
result_down = pd.concat([text_left_down, text_right_down] ,axis=1)
result_down.head()
```

```python
#然后将result-up和result-down，纵向合并为result
result = pd.concat([result_up, result_down], axis=0)
print(len(result_up))
print(len(result_down))
print(len(result))
```

```python
#5.1.4使用DataFrame自带的方法join和append，完成上述任务
result_up1 = text_left_up.join(text_right_up)
result_up1.head()
```

```python
result_down1 = text_left_down.join(text_right_down)
result_down1.head()
```

```python
result1 = result_up1.append(result_down1)
print(len(result_up1))
print(len(result_down1))
print(len(result1))
```

```python
#5.1.6 将数据保存为csv
result1.to_csv('data/result.csv')
```

### 6.1数据聚合与运算 ###

```python
#6.1.1 了解GroupBy

#参考链接：https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html

df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'], 
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
```

```python
gp = df.groupby(['key1'])
gp.groups
```

```python
gp.sum()
```

```python
#6.1.2计算泰坦尼克号男性与女性的平均票价
sex_fare_mean = text.groupby(['Sex'])['Fare'].mean()
sex_fare_mean
```

```python
#6.1.3统计泰坦尼克号男性与女性的存活人数
sex_surivived_count = text[text['Survived']==1].groupby(['Sex'])['Survived'].count()
sex_surivived_count
```

```python
#6.1.4计算客舱不同等级的存活人数
text[text['Survived']==1].groupby(['Pclass'])['Survived'].count()
```

```python
#6.1.5统计在不同等级的票中不同年龄的船票花费的平均值
text.groupby(['Pclass','Age'])['Fare'].mean()
```

```python
#6.1.6将任务二和任务三的数据合并，保存为csv
merge_grp = pd.concat([sex_fare_mean, sex_surivived_count], axis=1)
merge_grp
```

```python
merge_grp.to_csv('data/sex_fare_survived.csv')
```

```python
#6.1.7得出不同年龄总的存活人数
age_group = text['Survived'].groupby(text['Age']).sum()
age_group
```

```python
#然后找出存活人数的最高的年龄
max_survived_age = age_group[age_group.values==age_group.max()]
max_survived_age
```

```python
# 最后计算存活人数最高的存活率（存活人数／总人数）
age_group.max() / text['Survived'].sum()
```
