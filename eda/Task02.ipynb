补充① 缺失值&空值， 以及nan， naT, None的区别  
空值：在pandas中，空值就是空字符串 “”  
缺失值：np.nan（缺失数值），pd.naT（缺失时间），或None（缺失字符串）  
nan naT是numpy，pandas的缺失值类型，None是Python中的一种空值类型  
  
补充② 如何判断缺失值？如何查找筛选缺失值、空值？  
df.isna() 和 df.notna()，可以判断出None、pd.NaT、np.NaN 三种类型的缺失值；  
但是空字符串 “” 、0，不会认为是缺失值；另外对于4字符的字符串“None”、“null” ，也不会认为是缺失值。  
df.isnull() 等同于 df.isna()  
df.notnull() 等同于 df.notna()

### 4.4 重复值观察处理 ###

pandas.DataFrame.duplicated(self, subset=None, keep=’first’)  
subset：用于识别重复的列标签或列标签序列，默认所有列标签  
所以，不指定subset是默认所有列都相同则为重复(即完全相同的行则返回True)。  
keep:  
・first表示将重复行中的第一条视为false（即不标记为true）  
・last表示将重复行中的最后一条视为false（即不标记为true）  
・False表示将重复行全都标记为true

```python
#4.4.1查看数据中的重复值
df[df.duplicated()]
```

```python
#test
df.duplicated(subset=['Age'],keep = 'first').value_counts()
```

补充：python的直接赋值、深拷贝、浅拷贝  
直接赋值：其实就是对象的引用（别名）。  
浅拷贝(copy)：拷贝父对象，不会拷贝对象的内部的子对象。  
深拷贝(deepcopy)： copy 模块的 deepcopy 方法，完全拷贝了父对象及其子对象。

### 4.5 特征观察处理 ###

```python
#4.5.1 对年龄进行分箱处理
df['Age'].describe()
```

将连续型变量→离散型变量  

```python
# 将连续变量Age平均分箱成5个年龄段，并分别用类别变量12345表示
df['AgeBand'] = pd.cut(df['Age'], 5,labels=['1','2','3','4','5'])
df.head()
```

```python
#将连续变量Age划分为[0,5) [5,15) [15,30) [30,50) [50,80)五个年龄段，并分别用类别变量12345表示
df['AgeBand'] = pd.cut(df['Age'], [0, 5,15,30,50,80], right=False, labels=['1','2','3','4','5'])
df.head()
```

```python
#将连续变量Age按10% 30% 50 70% 90%五个年龄段，并用分类变量12345表示
# df['AgeBand'] = pd.qcut(df['Age'],[0,0.2,0.3,0.5,0.7,0.9],labels = ['1','2','3','4','5'])
df['AgeBand'] = pd.qcut(df['Age'].rank(method='first'),[0,0.1,0.3,0.5,0.7,0.9],labels = ['1','2','3','4','5'])
df.head()
```

```python
# 将文本变量Sex， Cabin ，Embarked用数值变量12345表示
#方法一，replace()
df['Sex_Num'] = df['Sex'].replace(['male','female'],[1,2])
df.head()
```

```python
#方法二：map（）
df['Sex_Num'] = df['Sex'].map({'male':1,'female':2})
df.head()
```

```python
#方法三: 使用sklearn.preprocessing的LabelEncoder
from sklearn.preprocessing import LabelEncoder
for feat in ['Cabin', 'Ticket']:
    lbl = LabelEncoder()
#     label_dict = dict(zip(df[feat].unique(), range(df[feat].nunique()) ))
#     df[feat + '_LabelEncoder'] = df[feat].map(label_dict)
    df[feat + '_LabelEncoder'] = lbl.fit_transform(df[feat].astype(str))
df.head()
```

```python
# 将文本变量Sex， Cabin， Embarked用one-hot编码表示a
```

```python
#将类别文本转换为one-hot编码
for feat in ['Age', 'Embarked']:
    x = pd.get_dummies(df[feat], prefix=feat )
    df = pd.concat([df, x], axis=1)
df.head()
```

```python
#4.5.3 从纯文本Name特征中提取Title特征（Mr,Miss,Mrs等）
df['Title'] = df.Name.str.extract('([A-Za-z]+)\.', expand=False)
df.head()
```
