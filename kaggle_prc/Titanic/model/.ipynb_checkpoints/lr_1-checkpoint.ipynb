##### 3.Cleanup and Transform the Data(Feature Engineering) #####

```python
#set missing age method-1

from sklearn.ensemble import RandomForestRegressor

def set_missing_age(df):
    age_df = df[['Age','Fare', 'Parch', 'SibSp', 'Pclass']]
    #  df.values　→ Return a Numpy representation of the DataFrame.(version 0.23: to_numpy() is recommended)
    #  print(age_df.values)
    # df.as_matrix()→ 将frame转换为它的数字数组表示形式(version 0.23.0: Use DataFrame.values() instead.)
    known_age = age_df[age_df.Age.notnull()].as_matrix()
    unknown_age = age_df[age_df.Age.isnull()].as_matrix()
    # y即目标年龄（第1列 age）
    y = known_age[:, 0]
    # X即特征属性值（第2列开始的其他列）
    X = known_age[:, 1:]
    
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
    rfr.fit(X, y)
    predictedAges = rfr.predict(unknown_age[:, 1:])
    
    # 用得到的预测结果填补原缺失数据
    df.loc[ (df.Age.isnull()), 'Age' ] = predictedAges 
    
    return df,rfr
```

```python
#set missing age method-2

# 根据多列的值做判断，生成新的列值
def set_sex_flg(name,sex):
    if 'Mrs' in name:
        sex_flg = 'Mrs'
    elif 'Miss' in name:
        sex_flg = 'Miss'
    elif 'Mr' in name:
        sex_flg = 'Mr' 
    else:
        if sex=='male':
            sex_flg='Mr'
        else:
            sex_flg = 'Miss'
    return sex_flg

def set_missing_age2(df):
    df_mean = df.groupby(df.sex_flg)['Age'].mean()
    print(df_mean)
    df_sex_flg = df.loc[(df.Age.isnull()), 'sex_flg'] 
    df.loc[(df.Age.isnull()), 'Age']  = df_sex_flg.apply(lambda x : df_mean.loc[x])
    
    return df
```

```python
#set missing cabin method-1

def set_cabin_type1(df):
    #找出notnull或isnull的行，在cabin列上设值yes或no
    df.loc[(df['Cabin'].notnull()), 'Cabin'] = "Yes"
    df.loc[(df['Cabin'].isnull()), 'Cabin'] = "No"
    return df
```

```python
#set missing cabin method-2

def get_cabin_no(s):
    cabinno = s[1:]
    if ' ' in cabinno:
        cabinno =  cabinno.split(sep=' ')[0]
    if ' ' in cabinno or cabinno=='':
            cabinno = 0
    return int(cabinno)

def set_cabin_type2(df):
    df['Cabin_1']=df.loc[df['Cabin'].notna(),'Cabin'].apply(lambda x: x[0:1])
    df['Cabin_2']=df.loc[df['Cabin'].notna(),'Cabin'].apply(lambda x: get_cabin_no(x))
    df.loc[(df['Cabin_2'].isnull()),'Cabin_2']=0
    return df
```

```python
#add child feature
def add_child(age):
    child_flg = 0
    if age <=12:
        child_flg= 1
    return child_flg
```

```python
# Sex&Pclass
def get_female_highclass(sex, pclass):
    flg = 0
    if sex=='female' and pclass!=3:
        flg = 1
    return flg

def get_male_lowclass(sex, pclass):
    flg = 0
    if sex=='male' and pclass==3:
        flg = 1
    return flg
```

```python
#add mother feature
def add_mother(sex, parch):
    mother_flg = 0
    if sex=='female' and parch>1:
        mother_flg =1
    return mother_flg
```

```python
# create title feature
def create_title_dict():
    title_dict ={}
    title_dict.update(dict.fromkeys(['Capt', 'Col', 'Major', 'Dr', 'Rev'],'Officer'))
    title_dict.update(dict.fromkeys(['Jonkheer', 'Don', 'Sir', 'the Countess', 'Dona', 'Lady'],'Royalty'))
    title_dict.update(dict.fromkeys(['Mme', 'Ms', 'Mrs'], 'Mrs'))
    title_dict.update(dict.fromkeys(['Mlle', 'Miss'], 'Miss'))
    title_dict.update(dict.fromkeys(['Mr'], 'Mr'))
    title_dict.update(dict.fromkeys(['Master'], 'Master'))
    print(title_dict)
    return title_dict
```

```python
#create family category 
def family_size_category(family_size):
    if family_size <= 1:
        return 'Single'
    elif family_size <= 3:
        return 'Small'
    else:
        return 'Large'
```

```python
def age_bins(age):
    if age<=12:
        return "Child"
    if age>12 and age<= 30:
        return "Youth"
    if age > 30 and age <= 50:
        return "Middle"
    else:
        return "Old"
```

```python
#将train和test数据合并到一起做特征处理
df_test['Survived'] = 0
df_combine=pd.concat([df_train,df_test])
df_combine
```

```python
# Fare缺失值填充
df_combine.loc[df_combine.Fare.isnull(), 'Fare'] = 0

df_combine['Group_Ticket'] = df_combine['Fare'].groupby(
    by=df_combine['Ticket']).transform('count')
df_combine['Fare'] = df_combine['Fare'] / df_combine['Group_Ticket']
df_combine.drop(['Group_Ticket'], axis=1, inplace=True)


def fare_category(fare):
    if fare <= 4:
        return 0
    elif fare <= 10:
        return 1
    elif fare <= 30:
        return 2
    elif fare <= 45:
        return 3
    else:
        return 4


df_combine['FareCategory'] = df_combine['Fare'].map(fare_category)
```

```python
#Age 缺失值填充
df_combine,rfr = set_missing_age(df_combine)
```

```python
#Cabin缺失值填充
df_combine = set_cabin_type1(df_combine)
```

```python
#Child 增加儿童特征，Age<1
df_combine['Child'] = df_combine['Age'].apply(lambda x: add_child(x))
```

```python
#Mother 增加母亲特征
df_combine['Mother'] = df_combine.apply(lambda x: add_mother(x.Sex,x.Parch), axis=1)
```

```python
#Sex&Pclass增加组合特征
df_combine['FemaleHighClass'] =  df_combine.apply(lambda x: get_female_highclass(x.Sex,x.Pclass), axis=1)
df_combine['MaleLowClass'] =  df_combine.apply(lambda x: get_male_lowclass(x.Sex,x.Pclass), axis=1)
```

```python
#Add Title Feature
# 第一次extract处理是：匹配正则表达式中的“，”的位置，并将“（）”括号中的部分抽出来，在下式中也就是“，” Name字段里“逗号”后面的部分截取留下来
#第二次extract处理是：匹配正则表达式中的“.”的位置，并将“（）”括号中的部分抽出来，在下式中也就是“.”Name字段里点前面的部分截取留下来
#（）代表匹配后被留下里的部分
df_combine['Title'] = df_combine['Name'].str.extract('.+,(.+)').str.extract( '^(.+?)\.').str.strip()
df_combine['Title'] = df_combine['Title'].map(create_title_dict())
```

```python
# 组合Parch与SibSp变为FamilyCategory
from sklearn import preprocessing
df_combine['FamilySize'] = df_combine['Parch'] + df_combine['SibSp'] + 1
df_combine['FamilyCategory'] = df_combine['FamilySize'].map(family_size_category)
```

```python
#Age 分段
# df_combine['Age'] = df_combine.Age.map(age_bins)
```

```python
#类别型特征，one-hot encoding
dummy_Cabin = pd.get_dummies(df_combine.Cabin,  prefix="Cabin")
dummy_Embarked = pd.get_dummies(df_combine['Embarked'],  prefix="Embarked")
dummy_Sex = pd.get_dummies(df_combine['Sex'], prefix="Sex")
dummy_Pclass =pd.get_dummies(df_combine['Pclass'], prefix="Pclass")
dummy_Title =pd.get_dummies(df_combine['Title'], prefix="Title")
dummy_FamilyCategory = pd.get_dummies(df_combine['FamilyCategory'],prefix="FamilyCategory")
dummy_FareCategory = pd.get_dummies(df_combine['FareCategory'],prefix="FareCategory")
# dummy_Age = pd.get_dummies(df_combine['Age'],prefix="Age")
```

```python
df_combine= pd.concat([df_combine, dummy_Cabin,dummy_Embarked,dummy_Sex,dummy_Pclass,dummy_Title,dummy_FamilyCategory,dummy_FareCategory]
                      , axis=1)
df_combine.shape
```

```python
#数值型特征，幅度缩放
import sklearn.preprocessing as preprocessing
scaler = preprocessing.StandardScaler()

df_combine['Age_scaled']= scaler.fit(df_combine[['Age']]) .transform(df_combine[['Age']])
# df_combine['Fare_scaled'] = scaler.fit(df_combine[['Fare']]).transform(df_combine[['Fare']])
```

##### 4.Split Training and Test Sets #####

```python
from sklearn import linear_model
import numpy as np

# y即Survival结果
y = df_train_new['Survived'].values

# X即特征属性值
X = df_train_new.drop(columns=['Survived']).values

#LogisticRegression
clf = linear_model.LogisticRegression(C=1.0, penalty='l1', tol=1e-6)
clf.fit(X, y)
clf

df_test_new.drop(columns=['Survived'],inplace=True)
predictions = clf.predict(df_test_new)
result = pd.DataFrame({'PassengerId':df_test['PassengerId'].as_matrix(), 'Survived':predictions.astype(np.int32)})
result.to_csv("../data/lr_predictions.csv", index=False)
```
