### 7.1导入 ###

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt## 第二章 数据重构##
```

```python
text = pd.read_csv('data/result.csv')
text.head()
```

### 7.2数据可视化 ###

```python
#7.2.1 了解matplotlib,创建一个数据项，对其进行可视化操作
```

折线图：常用于表示前后变化，对比，趋势信息  
条形图：对比，高低差距  
直方图：表现单独的维度信息  
箱线图：上边界，上四分位，中位数，下四分位，下边界  
散点图：体现相关程度  
饼状图：查看百分比  

Series对象的索引会被传给matplotlib，并用以绘制X轴。可以通过use_index=False禁用该功能。

```python
#7.2.2泰坦尼克号数据中男女生存人数分布状况
sex_group = text.groupby(['Sex'])['Survived'].sum()
sex_group
```

柱状图  
plot.bar()和plot.barh()分别绘制水平和垂直的柱状图。  
这时，Series和DataFrame的索引将会被用作X（bar）或Y（barh）刻度（如图9-15所示）

```python
sex_group.plot.bar()
plt.title('survived_count')
plt.show()
```

```python
#  7.2.3泰坦尼克号数据中男女生存与死亡人数比例
sex_survived = text.groupby(['Sex','Survived'])['Survived'].count()
sex_survived
```

```python
sex_survived = sex_survived.unstack()
sex_survived
```

```python
sex_survived.plot(kind='bar',stacked=True)
plt.title('survived_count')
plt.ylabel('count')
```

```python
#7.2.4 不同票价的人生存和死亡人数分布情况
fare_survived = text.groupby(['Fare'])['Survived'].value_counts().sort_values(ascending=False)
fare_survived
```

```python
fig = plt.figure(figsize=(20, 18)) 
fare_survived.plot(grid=True) 
plt.legend()
```

```python
fare_survived = text.groupby(['Fare'])['Survived'].value_counts()
fare_survived
```

```python
fig = plt.figure(figsize=(20, 18)) 
fare_survived.plot(grid=True) 
plt.legend()
```

```python
#7.2.5不同仓位的人生存和死亡人数分布情况
#方法一
import seaborn as sns
sns.countplot(x="Pclass", hue="Survived", data=text)
```

```python
#方法二
pclass_group = text.groupby(['Pclass'])['Survived'].value_counts().unstack()
pclass_group
```

```python
pclass_group.plot(kind='bar')
```

```python
#不同年龄的人生存和死亡人数分布情况
facet = sns.FacetGrid(text, hue="Survived",aspect=3)
facet.map(sns.kdeplot,'Age',shade= True)
facet.set(xlim=(0, text['Age'].max()))
facet.add_legend()
```

```python
#不同仓位的等级的人年龄分布状况
text[text['Pclass']==1]['Age'].plot(kind='kde')
text[text['Pclass']==2]['Age'].plot(kind='kde')
text[text['Pclass']==3]['Age'].plot(kind='kde')
plt.xlabel("age")
plt.legend((1,2,3),loc="best")
```
