### 2.2 筛选的逻辑###

```python
#2.2.1以Age为筛选逻辑，显示年龄在10岁以下的乘客信息
train[train.Age<10].head()
```

```python
#2.2.2以Age为条件，将年龄在10岁以上，和50岁以下的乘客信息显示出来，并命名为midage
midage = train[(train['Age']>10) & (train['Age']<50)]
midage.head()
```

```python
#2.2.3将midage数据中第100行的"Pclass"和"Sex"显示出来
midage.loc[100, ['Pclass', 'Sex']]
```

```python
#2.2.4使用loc方法将midage数据中第100，105，108行的 "Pclass" "Name" "Sex"数据显示出来
midage.loc[[100,105,108],['Pclass', 'Name', 'Sex']]
```

```python
#2.2.5使用iloc方法将midage数据中第100，105，108行的 "Pclass" "Name" "Sex"数据显示出来
midage.iloc[[100,105,108],[2, 3, 4]]
```

### 3.2 了解数据###

```python
#3.2.1利用Pandas对数据进行排序，升序
frame = pd.DataFrame(np.arange(8).reshape(2,4), index=['2', '1'], columns=['d', 'a', 'b', 'c'])
frame
```

```python
#按c列排序
frame.sort_values(['c'], ascending=False)
```

```python
#按行索引升序
frame.sort_index(axis=0)
```

```python
#按列索引升序
frame.sort_index(axis=1)
```

```python
#按列索引降序
frame.sort_index(axis=1, ascending=False)
```

```python
 #按两列数据同时降序排序
frame.sort_values(['a', 'c'], ascending=False)
```

```python
#3.2.2 对泰坦尼克号数据按票价和年龄两列进行降序排序
newtrain.sort_values(['票价', '年龄'], ascending=False)
```
