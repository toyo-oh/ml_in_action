##### Part3_Batch Training #####

```python
import matplotlib as plt
import numpy as np
import tensorflow as tf
sess = tf.Session()
```

```python
#np.random.normal→Draw random samples from a normal (Gaussian) distribution.
x_vals = np.random.normal(1,0.1,100)
#np.repeat→Repeat elements of an array.
y_vals = np.repeat(10., 100) 
x_data = tf.placeholder(tf.float32, shape=[None,1])
y_target = tf.placeholder(tf.float32, shape=[None,1])
A = tf.Variable(tf.random_normal(shape=[1,1]))
```

```python
my_output = tf.matmul(x_data, A)
```

```python
init_2 = tf.global_variables_initializer()
sess.run(init_2)
```

```python
#tf.reduce_mean→Computes the mean of elements across dimensions of a tensor
# If axis is None, all dimensions are reduced, and a tensor with a single element is returned.
m = my_output- y_target
#tf.square 对矩阵的每个元素都做平方
loss = tf.reduce_mean(tf.square(m))
```

```python
my_opt = tf.train.GradientDescentOptimizer(0.02)
train_step = my_opt.minimize(loss)
```

```python
batch_size = 20
```

```python
loss_batch=[]
for i in range(100):
    # Generates a random sample from a given 1-D array
    rand_index = np.random.choice(100, size=batch_size)
    # 20*1
    rand_x = np.transpose([x_vals[rand_index]])
    rand_y = np.transpose([y_vals[rand_index]])
    
    sess.run(train_step, feed_dict={x_data:rand_x, y_target:rand_y})
    
    if(i+1)%5==0:
        print('Step #'+str(i+1)+' A='+str(sess.run(A)))
        temp_m = sess.run(m, feed_dict={x_data:rand_x, y_target:rand_y})
        print('m='+str(temp_m))
        temp_loss =sess.run(loss, feed_dict={x_data:rand_x, y_target:rand_y})
        print('Loss='+str(temp_loss))
        loss_batch.append(temp_loss)
```
